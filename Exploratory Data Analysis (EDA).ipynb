```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  

import warnings
warnings.filterwarnings('ignore')
```

```python
data_df = pd.read_csv(r'https://raw.githubusercontent.com/Kamil128/ProjektPraktycznyRegresja/main/data/medical_cost/medical_cost.csv',)
```

```python
data_df.columns
```

**CORRELATION BETWEEN NUMERIC FEATURES**:
AGE, BMI, NUMBER OF CHILDREN, CHARGES
- NO HIGH CORRELATION TO BE SEEN

```python
plt.figure(figsize=(12,8))
count_corr=data_df.corr() 

sns.heatmap(count_corr)
```

**RELATIONSHIP BETWEEN SMOKING AND MEDICAL INSURANCE CHARGES PAID**
1. Smokers pay much more medical insurance charges than non-smokers

```python
sns.scatterplot(x="smoker", y="charges", data=data_df, size=(25), palette=sns.color_palette("Set2",10) )
```

**HOW AGE INFLUENCES NUMBER OF MEDICAL INSURANCE CHARGES PAID**
1. People pay more medical insurance charges with age
2. In general, men pay more charges with age than women
3. Smokers pay much more charges with age, compared to non-smokers

```python
sns.lmplot(x="age", y="charges", data=data_df, size=(8), hue="sex", palette=sns.color_palette("Set2",10))
```

```python
sns.lmplot(x="age", y="charges", data=data_df, hue="smoker", size=(8), palette=sns.color_palette("Set2",10))
```

**HOW BMI INFLUENCES NUMBER OF MEDICAL INSURANCE CHARGES PAID** 
1. People with higher BMI pay more medical insurance charges 
2. Men with high bmi pay more charges than women with high bmi
3. Smokers pay much more medical insurance charges no matter what bmi they have, compared to non-smokers, and they pay dramatically more insurance charges if their bmi is high 

```python
sns.lmplot(x="bmi", y="charges", data=data_df, size=(8), hue="sex", palette=sns.color_palette("Set2",10))
```

```python
sns.lmplot(x="bmi", y="charges", data=data_df, size=(8), hue="smoker", palette=sns.color_palette("Set2",10))
```

**HOW SMOKING/NON-SMOKING AND SEX INFLUENCE MEDICAL INSURANCE CHARGES PAID**
1. For non-smokers, on average women tend to pay slightly more medical insurance charges
2. For smokers, on average men tend to pay more medical insurance charges


```python
plt.figure(figsize=(10,6))
sns.barplot(x="smoker",y="charges", data=data_df, hue="sex", palette=sns.color_palette("Set2",10))
```
