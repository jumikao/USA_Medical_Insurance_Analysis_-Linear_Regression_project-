```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  

import warnings
warnings.filterwarnings('ignore')
```

```python
data_df = pd.read_csv(r'https://raw.githubusercontent.com/Kamil128/ProjektPraktycznyRegresja/main/data/medical_cost/medical_cost.csv',)
```

```python
data_df.columns
```

**CORRELATION BETWEEN NUMERIC FEATURES**:
AGE, BMI, NUMBER OF CHILDREN, CHARGES
- NO HIGH CORRELATION TO BE SEEN

```python
plt.figure(figsize=(12,8))
count_corr=data_df.corr() 

sns.heatmap(count_corr)
```

**HOW AGE INFLUENCES NUMBER OF MEDICAL INSURANCE CHARGES PAID**
1. People pay more medical insurance charges with age
2. In general, men pay more charges with age 
3. Smokers pay much more charges with age, compared to non-smokers

```python
sns.lmplot(x="age", y="charges", data=data_df, hue="sex", size=(8))
```

```python
sns.lmplot(x="age", y="charges", data=data_df, hue="smoker", size=(8))
```

**HOW BMI INFLUENCES NUMBER OF MEDICAL INSURANCE CHARGES PAID** 
1. People with higher BMI pay more medical insurance charges 
2. Men with high bmi pay more charges than women with high bmi
3. Smokers pay much more medical insurance charges no matter what bmi they have, compared to non-smokers, and they pay dramatically more insurance charges if their bmi is high 

```python
sns.lmplot(x="bmi", y="charges", data=data_df, size=(8), hue="sex")
```

```python
sns.lmplot(x="bmi", y="charges", data=data_df, size=(8), hue="smoker")
```

```python
sns.lmplot(x="age", y="charges", data=data_df, hue="sex", size=(8))
```
